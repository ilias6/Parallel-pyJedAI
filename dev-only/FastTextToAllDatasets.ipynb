```python
import pandas as pd

main_dir = '/home/gap2/Downloads/csvProfiles/'
```

```python
from nltk.stem import SnowballStemmer 

stemmer_ss = SnowballStemmer("english")   #an instance of Snowball Stemmer
```

```python
from nltk.corpus import stopwords
stop_words = set(stopwords.words('english'))
```

```python
from nltk.tokenize import word_tokenize

def cleanSentence(sentence):
    token_words=word_tokenize(sentence) #we need to tokenize the sentence or else stemming will return the entire sentence as is.
    stem_sentence=[]
    for word in token_words:
        if word not in stop_words:
            stem_sentence.append(stemmer_ss.stem(word))
            stem_sentence.append(" ") #adding a space so that we can join all the words at the end to form the sentence again.
    return "".join(stem_sentence)
```

```python
import sister

sentence_embedding = sister.MeanEmbedding(lang="en")
```

```python
# restaurants1 dataset
d1A = pd.read_csv(main_dir + 'D1partA.csv', sep='|', engine='python')
d1A
```

```python
d1A['Name'] = d1A['Name'].astype(str)
d1A['Aggregate Value'] = d1A['Aggregate Value'].astype(str)
d1A['Embedded Name'] = d1A.apply(lambda row : sentence_embedding(row['Name']), axis = 1)
d1A['Embedded Ag.Value'] = d1A.apply(lambda row : sentence_embedding(row['Aggregate Value']), axis = 1)
```

```python
d1A['Clean Name'] = d1A.apply(lambda row : cleanSentence(row['Name']), axis = 1)
d1A['Embedded Clean Name'] = d1A.apply(lambda row : sentence_embedding(row['Name']), axis = 1)
d1A['Clean Ag.Value'] = d1A.apply(lambda row : cleanSentence(row['Aggregate Value']), axis = 1)
d1A['Embedded Clean Ag.Value'] = d1A.apply(lambda row : cleanSentence(row['Aggregate Value']), axis = 1)
d1A
```
